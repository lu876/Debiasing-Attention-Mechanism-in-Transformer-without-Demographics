```python
import pandas as pd
training_set = pd.read_pickle('training.pkl')
test_set = pd.read_pickle('test.pkl')
```

```python
training_x = training_set.text
training_label = training_set.final_label
training_sensitive = training_set.final_target_category
```

```python
training_max_length = training_x.str.len().max()
print(training_max_length)
print(training_sensitive.value_counts())

for index, row in training_set.iterrows():
    if isinstance(training_x[index], list):
        training_x[index] = ' '.join(training_x[index])
        if training_label[index] == 'normal':
            training_label[index] = 0
        else:
            training_label[index] = 1
        if training_sensitive[index] is not None:
            if 'Women' in training_sensitive[index]:
                training_sensitive[index] = 1
            else:
                training_sensitive[index] = 0
                
        else:
            training_sensitive[index] = 0
            

            
                

print(training_set.final_label.value_counts())
print(training_sensitive.value_counts())
training_set = training_set.reset_index()
```
